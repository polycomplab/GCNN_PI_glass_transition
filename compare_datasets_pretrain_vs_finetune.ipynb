```python
import torch
from data.KGNNDataLoader import DataLoader
from data.datasets import ExperimentalDataset, SynteticDataset
```

```python
target_name = "Tg, K"

dataset_path = "datasets/PI_syn_subset_1K_for_finetune"
exp_dataset = ExperimentalDataset(root=dataset_path, target_name=target_name)
exp_loader = DataLoader(
    exp_dataset,
    batch_size=1,
    shuffle=False,
    num_workers=1)
```

```python
target_name = "Tg, K"

dataset_path = "datasets/PI_syn_subset_1K_for_pretrain"
syn_dataset = SynteticDataset(
    root=dataset_path,
    target_name=target_name)
syn_loader = DataLoader(
    syn_dataset,
    batch_size=1,
    shuffle=False,
    num_workers=1)
```

```python
exp_batch = next(iter(exp_loader))
syn_batch = next(iter(syn_loader))
```

```python

# assert len(exp_batch.keys) == len(syn_batch.keys), (len(exp_batch.keys), len(syn_batch.keys))
for attr_name, ev in exp_batch:
    print(ev)
    sv = syn_batch[attr_name]
    if attr_name == 'x':
        assert sv[:,:-1].equal(ev[:,:-1])  # excluding random features
    elif attr_name == 'edge_attr':
        continue  # random bond length can be swapped for some reason
    else:
        assert sv.equal(ev), attr_name
```
