```python
df = df.drop(columns=['Unnamed: 5', 'Unnamed: 6'])
```

```python
df.head()
```

```python
df_Askadsky = df[df['METHOD ID'] == 7]
df_Askadsky['SMILES'].size
```

```python
df_measured_tgs = df[df['METHOD ID'] != 7]
```

```python
df_measured_tgs['SMILES'].size
```

```python
df_measured_tgs_average = df_measured_tgs[['POLYMER ID', 'SMILES', 'Tg']].groupby(['POLYMER ID', 'SMILES'])['Tg'].mean().reset_index()
```

```python
df_measured_tgs_average
```

```python
df = df_measured_tgs_average
df.columns=["ID", "SMILES", "Tg, K"]
```

```python
df.to_csv("/polymers/GCNN_PI_glass_transition/datasets/PA_exp/raw/target_csv.csv")
```

```python
df_both = pd.merge(df_measured_tgs_average, df_Askadsky, on='SMILES')
```

```python
df_both
```

```python
# MAE
(df_both['Tg_x'] - df_both['Tg_y']).abs().mean()
```

```python
# relative error
((df_both['Tg_x'] - df_both['Tg_y']).abs() / df_both['Tg_x']).mean()
```

```python
# Sqrt of MSE denormalized
np.sqrt(((df_both['Tg_x'] - df_both['Tg_y'])**2).mean())
```
